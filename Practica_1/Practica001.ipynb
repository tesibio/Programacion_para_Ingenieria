## Operaciones booleanas
- - -
En python puedes implementar los operadores booleanos caracteristicos, solo que se utilizan palabras en Inglés en lugar de los simbolos tradicionales `&&`, `||`

### Operadores de comparación

* **Igualdad (==):** Comprueba si dos valores son iguales.
* **Desigualdad (!=):** Comprueba si dos valores no son iguales.
* **Mayor que (>):** Comprueba si el valor de la izquierda es mayor que el de la derecha.
* **Menor que (<):** Comprueba si el valor de la izquierda es menor que el de la derecha.
* **Mayor o igual que (>=):** Comprueba si el valor de la izquierda es mayor o igual que el de la derecha.
* **Menor o igual que (<=):** Comprueba si el valor de la izquierda es menor o igual que el de la derecha.

```python
a = 5
b = 10

#Evaluación de las operaciones lógicas
resultado1 = (a == b)
resultado2 = (a < b)
print(f"El resultado de (a==b) es: {resultado1}")
print(f"El resultado de (a<b) es: {resultado2}")
```

### Operadores lógicos
1. ***AND lógico (and):*** Devuelve True si ambas expresiones booleanas son True, de lo contrario, devuelve False.
2. ***OR lógico (or):*** Devuelve True si al menos una de las expresiones booleanas es True. Si ambas son False, devuelve False.
3. ***NOT lógico (not):*** Devuelve True si la expresión booleana es False y viceversa. Es un operador unario.

```python

t, f = True, False
print(f"t = {type(t)}, f = {type(f)}")

print(t and f) # Operación lógica "y";
print(t or f)  # Operación lógica "ó";
print(not t)   # Operación lógica "no";
print(t != f)  # Operación lógica "o exclusiva";
```

## Datos de tipo Fecha
- - -
Para poder trabajar con fechas en python, deberas de utilizar el modulo `datetime`

### Declaracion de fechas  y horas

```python
from datetime import datetime
from datetime import time

#Obtener la fecha y hora actual
fecha_actual = datetime.now()
print(f"Fecha y hora actual: {fecha_actual}")

#crear una fecha específica
fecha_personalizada = datetime(2023,10,21)
print(f"Fecha personalizada: {fecha_personalizada}")

#Obtener la hora actual
hora_actual = datetime.now().time()
print(f"Hora actual: {hora_actual}")

#Crear un objeto de tiempo específico
hora_específica = time(12,30,15)
print(f"Hora específica: {hora_específica}")
```

### Diferencia entre dos fechas:

```python
from datetime import datetime

fecha1 = datetime(2023,10,21)
fecha2 = datetime(2023,11,1)

diferencia = fecha2 - fecha1
print(f"Diferencia entre fechas: {diferencia}")
```

### Sumar o restar dias a una fecha:

```python
from datetime import datetime, timedelta

fecha = datetime(2023,10,21)
dias_a_sumar = 5

nueva_fecha = fecha + timedelta(days=dias_a_sumar)
print(f"Fecha original: {fecha}")
print(f"Nueva fecha despues de sumar {dias_a_sumar} días: {nueva_fecha}")
```

### Sumar o restar una cantidad de tiempo a una hora:

```python
from datetime import datetime, timedelta

hora_actual = datetime.now().time()
delta_tiempo = timedelta(hours = 2, minutes = 30)
nueva_hora = (
    datetime.combine(
        datetime.today(), hora_actual
        ) + delta_tiempo
    ).time()
print(f"La hora actual es: {hora_actual}")
print(f"Nueva hora despues de sumar 2 horas y 30 minutos: {nueva_hora}")
```

### Formatear fechas como cadenas de texto en específico:

```python
from datetime import datetime

fecha = datetime(2023,10,21)
cadena_fecha = fecha.strftime("%Y-%m-%d")
print(f"Cadena formateada de la fecha: {cadena_fecha}")
```

## Datos de tipo categrico
- - -
En Python, la manera más rápida de manejar datos de tipo categórico es utilizando la "biblioteca" `pandas` para trabajar con datos categóricos

la libreria de `pandas` ofrece una funcionalidad muy completa para trabajar con datos categóricos y es ampliamente utilizada en el análisis de datos en Python. En caso de que **NO** tengas instalado pandas en tu entorno de Python, antes de ejecutar los siguientes ejemplos ejecuta la siguiente linea en la terminal

`pip install pandas`

```python
#Importamos la libreria pandas
import pandas as pd

#Creamos un arreglo de datos categóricos
categorias = ['Rojo', 'Verde', 'Azul']
datos = pd.Categorical(
    ['Rojo', 'Verde', 'Azul', 'Rojo'],
    categories = categorias
)

#Mostramos en pantalla el arreglo categórico
print(datos)
```

### Comparacion y filtrado

```python
rojos = datos == 'Rojo'
print(rojos)
```

### Frecuencia de categorías

```python
frecuencia = datos.value_counts()
print(frecuencia)
```
