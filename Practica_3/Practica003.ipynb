## Declaración de funciones
- - -
La definicion de funciones en Python es realizada mediante la palabra clave `def`. Por ejemplo



```python
def sign(x):
    if x > 0:
        return 'Valor Positivo'
    elif x < 0:
        return 'Valor negativo'
    else:
        return 'cero'

for x in [-1, 0, 1]:
    print(sign(x))
```

Estas funciones pueden definirse con argumentos opcionales directamente en la firma de la funcion y asignarles valores predeterminados. De esta manera, los argumentos son opcionales y permiten proporcionar valores específicos cuando se llama a la funcion, pero si no se proporciona un valor se utilizara el valor predeterminado

```python
def REGISTRO_PERSONA(Nombre = None, Edad = None, Ciudad = None):
    if Nombre is not None:
        print(f"Nombre: {Nombre}")
    if Edad is not None:
        print(f"Edad: {Edad}")
    if Ciudad is not None:
        print(f"Ciudad: {Ciudad}")

#Llamada a la funcion
REGISTRO_PERSONA(Nombre = "Juan", Ciudad = "Jalisco")
```

Como puedes ver, en este ejemplo `Nombre`, `Edad` y `ciudad` son los argumentos de entrada de la funcion `REGISTRO_PERSONA`. Estos argumentos son opcionales, por lo que se les asigna un valor predeterminado `None`. Dentro de esta función se verifica si cada argumento tiene un valor diferente de `None` antes de mostrarlo en pantalla.


### **Kwargs:
Existe otra manera de declarar argumentos de entrada de la funcion `REGISTRO_PERSONA`. Esta es mediante la sintaxis `**kwargs`. Esta sintaxis permite pasar un número variable de argumentos de palabras clave, para que la funcion los recoja en un diccionario.

```python
def REGISTRO_PERSONAV2(**kwargs):
    if 'nombre' in kwargs:
        print(f"Nombre: {kwargs['nombre']}")
    if 'edad' in kwargs:
        print(f"Edad: {kwargs['edad']}")
    if 'ciudad' in kwargs:
        print(f"Ciudad: {kwargs['ciudad']}")

#Llamada a la función con argumentos opcionales
REGISTRO_PERSONAV2(nombre = "Juan", ciudad = "Jalisco")
```

En este ejemplo, la función `REGISTRO_PERSONAV2` acepta argumentos de palabras clave opcionales. La funcion verifica si cada clave específica esta presente en el diccionario `kwargs`. Si es así se imprime el valor correspondiente

Al llamar a la funcion con los argumentos:
`nombre = Juan` y `ciudad = Jalisco`, solo se imprimen las partes relevantes del diccionario

### Variables globales: 

Las variables globales son aquellas que se definen fuera de una función y estan dsponibles en todo el script o módulo. Estas se pueden acceder desde cualquier parte del código, incluidas las funciones. Para modificar una variable global desde una función, debes declararla como `global` dentro de la función

```python
variable_global = 10

def funcion_con_global():
    global variable_global
    variable_global += 5
    return variable_global

resultado = funcion_con_global()
print(resultado)
```

### Variables Locales:

Las variables locales se definen dentro de una función y solo son accesibles dentro de esa función. No afectan a variables con el mismo nombre fuera de la función

```python
def funcion_con_local():
    variable_local = 10
    variable_local += 5
    return variable_local

resultado = funcion_con_local()
print(resultado)

#Esto generará un error, ya que la variable "variable_local" no esta definida aquí
print(variable_local)
```

### Variables no locales (`nonlocal`)

En funciones anidades, puedes utilizar la palabra clave `nonlocal` para referirte a una variable de un ambito externo al inmediato, pero no global. Permite modificar una variable en el ámbito de la función más externa

```python
def funcion_externa():
    variable_externa = 30
    
    def funcion_interna():
        nonlocal variable_externa
        variable_externa += 5
        
    funcion_interna()
    return variable_externa

resultado = funcion_externa()
print(resultado)
```

## Clases
- - -
La Programación Orientada a Objetos (POO) es un paradigma de programación que organiza el código alrededor de "objetos". Estos "objetos" Son instancias de clases. Cada objeto tiene atributos (caracteristicas) y métodos (funciones) asociados.
Los principales conceptos en los que se basa la POO son


1.   **Clase:** Plantilla para crear objetos. Define atributos y métodos comunes
2.   **Objeto:** Instancia de una clase con atributos y métodos específicos.
3.   **Atributo:** Caracteristica de un objeto que almacena información
4.   **Método:** Función asociada a una clase que opera en sus instancias
5.   **Constructor:** Método especial que se llama al crear un objeto e inicializa sus atributos.

En Python, un **constructor** es un método especial en una clase que se llama automáticamente cuando se crea una instancia (un objeto). Este es responsable de inicializar los atributos del objeto. El constructor se llama `__init__`

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
        print(nombre)
        print(edad)

#Creacion de la instancia (objeto) de la clase Persona
#Instancia 1
P1 = Persona(
    nombre = "Juan",
    edad = 25
)

#Instancia 2
P2 = Persona(
    nombre = "Maria",
    edad = 32
)
```

En este ejemplo `__init__` es el _constructor_ de la **clase** `Persona`. Cuando se crea una *instancia de la clase(objeto)*, en este caso P1, se llama automáticamente con los argumentos proporcionados.

Una **instancia** es un objeto específico creado a partir de una clase. En el ejemplo anterior, `P1` es una instancia de la clase `Persona`. Las instancias tienen atributos específicos que pueden ser diferentes para cada instancia, pero tienen en comun que comparten la misma estructura definida por la clase

Por su parte, un **método** es una función asociada a una clase. En Python, los métodos se definen dentro de la clase y tienen un primer parámetro especial llamado `self`que se refiere a la instancia actual de la clase

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def saludar(self):
        print(f"Hola, mi nombre es {self.nombre} y tengo {self.edad} años")

#Creamos la primera instancia de la clase persona
P1 = Persona(
    nombre = "Juan",
    edad = 25
)

#Creamos la segunda instancia de la clase persona
P2 = Persona(
    nombre = "Maria",
    edad = 32
)

#Finalmente llamamos al método saludar, el cual es perteneciente a la instancia P1
P1.saludar()

#Podemos llamar al método saludar pero aquel perteneciente a la instancia P2
P2.saludar()
```

En este ejemplo, saludar es un método de la clase `Persona`. Cuando se llama `juan.saludar()`, el método se ejecuta con la instancia `juan`, por lo que puede acceder a los atributos específicos de dicha instancia.

En conclusión, la POO permite encapsular datos y comportamientos relacionados dentro de un solo objeto (`Persona` en este caso), facilitando la organización y reutilización del código. Cada instancia de la clase tiene sus propios datos *(atributos)* y puede realizar acciones específicas *(metodos)*

## Ciclos de programación (Loops)
- - -
En Python, los ciclos de programacion dentro de las listas son simplificados para "favorecer" la legibilidad de escritura, lectura y lógica del código. 

### Ciclo de programación `if`


```python
numero = float(input("Ingresa un numero: "))

if numero > 0:
    print("El número es positivo.")
elif numero < 0:
    print("El número es negativo")
else:
    print("El número es cero.")
```

### Ciclo de programación `for`

```python
#Primero se declara una lista que tiene en su interior 3 datos de tipo cadena de texto
animales = ['gato', 'perro', 'mono']
print (animales)

#Observa que si queremos acceder al segundo elemento de la lista tendriamos que declarar:
print(animales[1])

#Por otra parte, si queremos recorrer todos los elementos dentro de la lista de uno a uno,
#podemos declarar un ciclo for de la siguiente manera

for animal in animales:
    print(animal)
```

El bucle `for` declarado dentro del lenguaje de Python se encarga automáticamente de asignar a la variable `animal` cada elemento de la lista llamada `animales` en cada iteración. Esta asociación no se raliza a través de un índice numérico como en algunos otros lenguajes, sino que la variable `animal` toma directamente el valor de cada elemento en la lista


```python
#En caso de que sea necesario, puedes utilizar el comando enumerate() para obtener el índice junto con el valor
for index, animal in enumerate(animales):
    print(f'Índice: {index}, Animal: {animal}')
```

### Ciclo de programación `switch`
En Python, no hay una construcción de control de flujo llamado `switch` como en algunos otros lenguajes de programación. Sin embargo, puedes lograr un comportamiento similar utilizando una estructura de datos de tipo diccionario

```python
def case1():
    return "Ejecutando caso 1"

def case2():
    return "Ejecutando caso 2"

def case3():
    return "Ejecutando caso 3"

def default():
    return "Caso no válido"

#Se define el diccionario que apunta a las funciones
switch_dict = {
    1: case1,
    2: case2,
    3: case3
}

#A partir de ahora se inicia el ciclo switch
opcion = int(input("Ingresa una opción (1, 2, o 3):"))

#Obtiene y se ejecuta la función correspondiente al caso
resultado = switch_dict.get(opcion, default)()

print(resultado)
```

### Ciclo de programación `while`

```python
import random

#Primero se genera un número aleatorio entre 1 y 10
numero_secreto = random.randint(1, 10)

#Se inicializa la variable de intentos
intentos = 0

#Se realiza el ciclo while para adivinar el número
while True:
    #Solicitar al usuario que ingrese un número
    adivinanza = int(input("Adivina el número (entre 1 y 10):"))
    
    #Incrementa el contador de intentos
    intentos += 1
    
    #Verifica si la adivinanza es correcta
    if adivinanza == numero_secreto:
        print(f"Correcto! Has adivinado el número en {intentos} intentos.")
        break
    else: 
        print("Incorrecto. Intentalo de nuevo,")
```
